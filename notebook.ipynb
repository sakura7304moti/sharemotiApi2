```python
import yt_dlp

url = 'https://www.youtube.com/channel/UCvaTdHTWBGv3MKj3KVqJVCw'

def get_yt_info(url:str):
    # オプションを設定
    ydl_opts = {
        'quiet': True,  # 出力を非表示
        'extract_flat': True,  # プレイリスト内の各動画を一つのリストに展開
        'date':True,
        'output':'%(upload_date)'
        # その他のオプションは必要に応じて設定
    }
    
    # yt-dlpオブジェクトを生成
    ydl = yt_dlp.YoutubeDL(ydl_opts)
    
    # プレイリストからURLリストを取得
    with ydl:
        result = ydl.extract_info(url, download=False)
    return result
```

```python
"""
チャンネルデータ
"""
class ChannelInfo:
    def __init__(self,channel_name,description,header_url,avatar_url):
        self.channel_name = channel_name
        self.description = description
        self.header_url = header_url
        self.avatar_url = avatar_url
        
    def __str__(self):
        return f"""channel_name -> {self.channel_name}
header_url -> {self.header_url}
avatar_url -> {self.avatar_url}
---- description -----
{self.description}
        """

def get_channel_info(result):
    #チャンネル名
    channel_name = result['channel']
    
    #概要
    description = result['description']
    
    
    
    #ヘッダー画像URL
    for rec in result['thumbnails']:
        if 'resolution' in rec:
            if rec['resolution'] == '1920x1080':
                header_url = rec['url']
                
    #プロフィール画像URL
    avatar_url = result['thumbnails'][-2]['url']
    return ChannelInfo(channel_name,description,header_url,avatar_url)

channel_info = get_channel_info(result)
print(channel_info)
```

```python
import urllib
"""
動画データ
"""
class MovieInfo:
    def __init__(self,index,id,url,title,view_count,thumbnail_url):
        self.index = index
        self.id = id
        self.url = url
        self.title = title
        self.view_count = view_count
        self.thumbnail_url = thumbnail_url

    def __str__(self):
        return f"""index -> {str(self.index).zfill(4)}
if -> {self.id}
url -> {self.url}
title -> {self.title}
view_count -> {self.view_count}
thumbnail_url -> {self.thumbnail_url}"""

def get_movie_info(entrie):
    id = entrie['id']
    url = entrie['url']
    title = entrie['title']
    if 'view_count' in entrie:
        view_count = entrie['view_count']
        thumbnail_url = get_thumbnail_url(id)
        return [id,url,title,view_count,thumbnail_url]
    else:
        return None

def get_thumbnail_url(id):
    return f'http://img.youtube.com/vi/{id}/maxresdefault.jpg'

def get_movie_list(entries):
    movie_list = []
    for index,rec in enumerate(entries):
        info = get_movie_info(rec)
        if info is not None:
            movie_index = abs(index - len(entries))
            movie_info = MovieInfo(movie_index,*info)
            movie_list.append(movie_info)
    return movie_list
    
#動画
movie_list = get_movie_list(result['entries'][0]['entries'])

#ライブ
live_list = get_movie_list(result['entries'][1]['entries'])

#ショート
short_list = get_movie_list(result['entries'][2]['entries'])
```

```python
import re
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome import service as fs
from tqdm.notebook import tqdm

url = 'https://hololive.hololivepro.com/talents'

from selenium.webdriver.chrome.options import Options

options = Options()
options.add_argument('--headless')
driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()),options=options)
driver.get(url)
```

```python
#無料キャラ
xpath = '/html/body/div[2]/main/div[2]/div/ul'

ul = driver.find_element(by=By.XPATH, value=xpath)
li_list = ul.find_elements(by=By.TAG_NAME,value='li')

url_list = []
for li in li_list:
    url = li.find_element(by=By.TAG_NAME,value='a').get_attribute("href")
    url_list.append(url)
        
for url in url_list:
    print(url)
```

```python
channel_list = []
for url in tqdm(url_list):
    driver.get(url)
    a = driver.find_element(By.XPATH,'/html/body/div[2]/main/article/div[1]/div[1]/div[2]/div/ul/li[1]/a')
    channel_url = a.get_attribute('href').replace('?sub_confirmation=1','')
    channel_list.append(channel_url)

for url in channel_list:
    print(url)
```

```python
import pandas as pd

df = pd.Series(channel_list)
df.to_csv('channel_list.csv',index=False)
```

```python
import yt_dlp
import datetime

def get_upload_date(url:str):
    # オプションを設定
    ydl_opts = {
        'quiet': True,  # 出力を非表示
        'date':True,
        'output':'%(upload_date)'
        # その他のオプションは必要に応じて設定
    }
    
    # yt-dlpオブジェクトを生成
    ydl = yt_dlp.YoutubeDL(ydl_opts)
    
    # プレイリストからURLリストを取得
    with ydl:
        result = ydl.extract_info(url, download=False)
        stamp = result['release_timestamp']
        return str(datetime.datetime.fromtimestamp(stamp))
```

## スクレイピング修正

```python
import requests
from bs4 import BeautifulSoup
import re

def get_original_songs():
    records = []
    response = requests.get('https://seesaawiki.jp/hololivetv/d/%a5%aa%a5%ea%a5%b8%a5%ca%a5%eb%a5%bd%a5%f3%a5%b0')
    soup = BeautifulSoup(response.text, 'html.parser')
    elements = soup.find_all('div',class_="wiki-section-3")
    for element in elements:
        text = element.text
        title = text.splitlines()[0]
        member_match = re.search(r"メンバー：(.+?)\n", text)
        if member_match:
            members = member_match.group(1)
        date_match = re.search(r"音源公開日：(\d{4}/\d{2}/\d{2})\n", text)
        if date_match:
            release_date = date_match.group(1).replace('/','-')
        a_tags = element.find_all('a', {'class': 'outlink', 'href': True, 'target': '_blank'})
        youtube_link = [a for a in a_tags if 'youtube' in a.text.lower()]
        if len(youtube_link) > 0:
            youtube_url = youtube_link[0]['href']
            rec = (release_date,members,youtube_url,title,'')
            if youtube_url != 'https://www.youtube.com/watch?v=VKyEVCPEV6c':
                records.append(rec)
    return records

records = get_original_songs()

for rec in records:
    print(rec)
```
